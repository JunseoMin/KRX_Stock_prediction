```python
import pandas as pd

data=pd.read_csv('/home/junseo/Desktop/2023_1st_vacation/KRX_py/KRX_DATA/open/train.csv')
res=pd.read_csv('~/Downloads/230723_v0.1.csv')

print(data)

```

```python
data=data.sort_values(['일자','종목코드'])
```

```python
end=data['종가'].iloc[-2001:-1]
end
befoe_price=[]
for b_p in end:
    befoe_price.append(b_p)
befoe_price
```

```python
end_price=[]
ed_q=res['종가']
ed_q
for ed_p in ed_q:
    end_price.append(ed_p)
end_price
```

```python
idx=0
percent=[]

for value in end_price:
    percent.append((value-befoe_price[idx])/befoe_price[idx]*100)
    
percent
```

```python
res['percent']=percent
```

```python
res
```

```python
res.sort_values('percent',ascending=False,inplace=True)
res
```

```python
index=[i for i in range(1,2001)]
index[-1]
```

```python
res['순위']=index
res
```

```python
res.drop(['Unnamed: 0','종가','percent'],axis=1,inplace=True)
```
